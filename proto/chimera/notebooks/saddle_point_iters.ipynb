```python
%matplotlib inline

import matplotlib.pylab as plt
import numpy as np
import scipy.special as sp
```

```python
m = 20

p_lo = np.random.rand(m)
p_hi = np.random.rand(m)

p_lo = p_lo / np.sum(p_lo)
p_hi = p_hi / np.sum(p_hi)

e_lo = np.random.randint(1, 10)
e_hi = np.random.randint(1, 10)
e_tot = e_lo + e_hi

n_r = np.random.multinomial(e_lo, p_lo) + np.random.multinomial(e_hi, p_hi)
```

```python
plt.plot(n_r / np.sum(n_r))
plt.plot(e_lo * p_lo / e_tot + e_hi * p_hi / e_tot)
```
